```python
columns_to_encode = [
    'Marital status',
    'Application mode',
    'Application order',
    'Course',
    'Previous qualification',
    'Nacionality',
    "Mother's qualification",
    "Father's qualification",
    "Mother's occupation",
    "Father's occupation"
]

df[columns_to_encode] = df[columns_to_encode].astype(str)

df = pd.get_dummies(df, columns=columns_to_encode, drop_first=True, dtype=int)
```

```python
df.columns = df.columns.str.strip()
```

```python
bool_columns = [
    'Daytime/evening attendance',
    'Displaced',
    'Educational special needs',
    'Debtor',
    'Tuition fees up to date',
    'Scholarship holder',
    'International'
]
```

```python
mapping = {'yes': 1, 'no': 0}
```

```python
for col in bool_columns:
    if col in df.columns:
        df[col] = df[col].replace(mapping)
        df[col] = pd.to_numeric(df[col]).astype(int)
```

```python
df[bool_columns].head()
```

```python
numerical_columns = [
    'Previous qualification (grade)',
    'Admission grade',
    'Age at enrollment',
    'Curricular units 1st sem (credited)',
    'Curricular units 1st sem (enrolled)',
    'Curricular units 1st sem (evaluations)',
    'Curricular units 1st sem (approved)',
    'Curricular units 1st sem (grade)',
    'Curricular units 1st sem (without evaluations)',
    'Curricular units 2nd sem (credited)',
    'Curricular units 2nd sem (enrolled)',
    'Curricular units 2nd sem (evaluations)',
    'Curricular units 2nd sem (approved)',
    'Curricular units 2nd sem (grade)',
    'Curricular units 2nd sem (without evaluations)',
    'Unemployment rate',
    'Inflation rate',
    'GDP'
]
```

```python
cols_to_scale = [col for col in numerical_columns if col in df.columns]
```

```python
df[cols_to_scale] = df[cols_to_scale].astype(float)
```

```python
scaler = StandardScaler()
```

```python
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
```
